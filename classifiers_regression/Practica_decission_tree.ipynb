```python
dfl = df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2',
               'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
               'CLIENTNUM'],axis=1)
```

```python
dfl.Card_Category = df.Card_Category.map({'Blue': 1, 'Gold': 2, 'Silver': 3, 'Platinum': 4})
dfl.Income_Category = df.Income_Category.map({'$60K - $80K' : 1, 'Less than $40K' : 2, '$80K - $120K' : 3, '$40K - $60K': 4,
       '$120K +': 5, 'Unknown' : 6})
dfl.Marital_Status = df.Marital_Status.map({'Married': 1, 'Single': 2, 'Unknown': 3, 'Divorced': 4})
dfl.Education_Level = df.Education_Level.map({'High School': 1, 'Graduate': 2, 'Uneducated' : 3, 'Unknown': 4, 'College' : 5,
       'Post-Graduate': 6, 'Doctorate': 7})
dfl.Gender = df.Gender.map({'M' : 1, 'F': 2})
dfl.Attrition_Flag = df.Attrition_Flag.map({'Existing Customer':1, 'Attrited Customer':2})
```
